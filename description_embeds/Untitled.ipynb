```python
import pickle
import torch
from nltk.corpus import wordnet
```

```python
with open("miniImageNet_bert-base-cased_layer6.pickle", 'rb') as f:
    d = pickle.load(f)
```

```python
vocab = list(d.keys())
```

```python
defs = [wordnet.synsets(v.replace(" ", "_"))[0].definition() for v in vocab]
```

```python
for i,df in enumerate(defs):
    print(f"{vocab[i]}: {df}")
```
